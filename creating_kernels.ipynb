# Kernels
A kernel is the engine that will execute your code. We can install packages directly from the notebook and it will be attached to your kernel. If you whish to create a new kernel, below are instructions on how to create new kernels. However, you could also use the default Python kernel and install packages there. By default, there is no R kernel available for Jupyter, we can build one though.
In general, R-studio is easier to use than jupyter notebook as it's orientation is Python. Below are instructions on how to create and install packages for Python and R. 

## Installing packages from the notebook Python only
In a notebook cell, the following code will allow you to install packages. Restart the kernel after installing packages to ensure you can access them from your notebook.
```python
%pip install <mypackage>
```

## Creating your own Python and R kernels
Below are a list of commands per programming language. Execute these commands in terminal (bash).
It takes a few minutes before the jupyter launcher recognizes the updates kernels. Give it a few minutes (2-5) for it to update. 

### Python
#### Create a conda environment for your kernel and activate it
```bash
conda create -n pyocean -c conda-forge ipykernel
conda activate pyocean
```
#### Create a kernel out of the new conda environment
```
python -m ipykernel install --user --name pyocean --display-name "Py-Ocean"
```

#### Install a package
We can do this in Conda or in the notebook through pip. In general, its easier to install in the notebook through pip. See [Installing packages from the notebook directly (Python Only)](#installing-packages-from-the-notebook-python-only)
However, if you have a requirement to use Conda, you can do so by either opening a terminal and entering the code below.
Minio is required to communicate with Abyssio server
```bash
conda install -c conda-forge minio
```

### R
#### Create a conda environment for your kernel and activate it
```bash
conda create -n rocean -c conda-forge r-essentials
conda activate rocean
```

#### Start R
```bash
R
```

#### Install packages in R
The IRkernel package is required to build a kernel, aws.s3 is required to communicate with Abyssio server.
```R
install.packages(c('aws.s3','IRkernel'))
```

#### Update kernel
```R
IRkernel::installspec(name = 'rocean', displayname = 'R-Ocean')
```
#### Exit R
```R
exit()
```
